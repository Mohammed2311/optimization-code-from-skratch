```python
x,y = symbols('x y')
f = x**2 * y
plot3d(f,size=(10,7))
plt.show()
```

```python
f_1 = lambda x: x**2  + y**2
Df_1 = lambda X: np.array([2*X[0],2*X[1]])
DDf_1 = lambda X: np.array([[2,0],
                         [0,2]])
```

```python
f_2 = lambda x: x**2 + x*y + y**2
Df_2 = lambda X: np.array([2*X[0]+X[1],2*X[1]+X[0]])
DDf_2 = lambda X: np.array([[2,1],
                        [1,2]])
```

```python
# First function
# Multi-variable Gradient descent with learning rate = 0.1,
# x0=(10,10), epsilon = 0.001

x =[10, 10]
e = 0.001
lr= 0.1
df1 = Df_1(x)
df2 = DDf_1(x)

for i in range(1000):
    x = x -lr*df1
    df1 = Df_1(x)
    df2 =DDf_1(x)
    nr = np.linalg.norm(df1)
    if nr < e:
        break
    
    
print(f"Found solution after {i+1} iterations.")
print(f"xy_min = {x}")
print(f"Gradient = {df1}")
```

```python
# First function
# Multi-variable Gradient descent with learning rate = 0.5,
# x0=(10,10), epsilon = 0.001
x =[10, 10]
e = 0.001
lr= 0.5
df1 = Df_1(x)
df2 = DDf_1(x)

for i in range(1000):
    x = x -lr*df1
    df1 = Df_1(x)
    df2 =DDf_1(x)
    nr = np.linalg.norm(df1)
    if nr < e:
        break

print(f"Found solution after {i+1} iterations.")
print(f"xy_min = {x}")
print(f"Gradient = {df1}")
```

```python
def newton_method(x , e ,lr , iteration  ):
    df1=Df_1(x)
    df2 =DDf_1(x)
    for i in range(iteration):
        x = x - lr *(np.linalg.inv(df2)@df1)
        df1 = Df_1(x)
        df2 = DDf_1(x)
        if np.linalg.norm(df1) <e :
            break
    print(f'after {i+1} iterations ' )
    print(f'x_min =  {x}')
    print(f'Gradient ={ df1}')

```

```python
# First function
# Multi-variable Newton's method without learning rate, 
# x0=(10,10), epsilon = 0.001
print("Multi-variable Newton's method without learning rate \n")
newton_method([10,10] ,0.001 ,1 ,1000) 
```

```python
# First function
# Multi-variable Newton's method with learning rate = 0.5, 
# x0=(10,10), epsilon = 0.001
print("Multi-variable Newton's method with learning rate = 0.5\n")
newton_method([10,10] ,0.001 ,0.5 ,1000) 
```

## BFGS
- Implement the <b>BFGS</b> as a function and find the minimum of the following function:
    - $ùíá(ùíô)=ùíô^ùüê+y^2$
    - $ùíá(ùíô)=ùíô^ùüê+xy+y^2$  
    - $ùíá(ùíô)=ùíô^ùüêy$ 
- The stop condition is grad < epsilon.
- Try different epsilons and initial conditions.
- Compare the results from Gradient descent and Newton's method.

```python

I= np.eye(2)
x0 =np.array(  [10,10]).reshape((-1,1))
x1 =np.array([9,9]).reshape((-1,1))
x=[]
x.append(x0)
x.append(x1)
a = 1
B = np.eye(2,2)
B1 =[]
B1.append(B)
i = 0
for i in range(1000):
    delta_x = x[-1]- x[-2]
    y = Df_1(x[-1])-Df_1(x[-2])
    first_term = I  - (delta_x @y.T)/y.T @delta_x
    second = I - ((y@delta_x.T )/(y.T@delta_x))
    B = (first_term )@B @(second ) +(delta_x@delta_x.T)/(y.T @delta_x)
    B1.append(B)
    x1 = x[-1] - a*(B @Df_1(x[-1]))
    x.append(x1)
    if np.linalg.norm(Df_1(x[-1]))<0.001:
        break
        

print(i)
print(B1[-1])
print(x[-1])

    
    
```
